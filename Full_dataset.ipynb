## Reformat acolite datasets so have range instead of min and max and remove point data

```python
#refine acolite datasets to only include area data and turn max and min into range

#select only those with area
ac_max = ac_max.filter(like='area')
ac_mean = ac_mean.filter(like='area')
ac_min = ac_min.filter(like='area')
ac_sum = ac_sum.filter(like='area')
```

```python
ac_max.keys()
```

```python
#join ac_max and ac_min by index
ac_range = pd.merge(ac_max, ac_min, left_index = True, right_index = True)
#k = ac_range.keys().tolist()
#subtract max and min to get range
```

```python
ac_range.keys()
```

```python
names = ['chl_oc3',
 'chl_re_gons',
 'TUR_Nechad2016_665',
 'SPM_Nechad2016_665',
 'p3qaa_zSD',
 'p3qaa_zSD_biased',
 'p3qaa_eta',
 'p3qaa_KPAR']

cols = ['range_' + e for e in names]
cols
```

```python
#calculate the ranges as new column
ac_range['range_chl_oc3'] = ac_max['max_area_chl_oc3'] - ac_min['min_area_chl_oc3']
ac_range['range_chl_re_gons'] = ac_max['max_area_chl_re_gons'] - ac_min['min_area_chl_re_gons']
ac_range['range_TUR_Nechad2016_665'] = ac_max['max_area_TUR_Nechad2016_665'] - ac_min['min_area_TUR_Nechad2016_665']
ac_range['range_SPM_Nechad2016_665'] = ac_max['max_area_SPM_Nechad2016_665'] - ac_min['min_area_SPM_Nechad2016_665']
ac_range['range_p3qaa_zSD'] = ac_max['max_area_p3qaa_zSD'] - ac_min['min_area_p3qaa_zSD']
ac_range['range_p3qaa_zSD_biased'] = ac_max['max_area_p3qaa_zSD_biased'] - ac_min['min_area_p3qaa_zSD_biased']
ac_range['range_p3qaa_eta'] = ac_max['max_area_p3qaa_eta'] - ac_min['min_area_p3qaa_eta']
ac_range['range_p3qaa_KPAR'] = ac_max['max_area_p3qaa_KPAR'] - ac_min['min_area_p3qaa_KPAR']
```

```python
#only keep those columns with 'range' - remove the min and max columns
all_cols = ac_range.keys().tolist
keep_cols = [col for col in ac_range if col.startswith('range')]
ac_range = ac_range[keep_cols]
```
